importing libraries and reading data from the dataset 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

Reading the dataset and viewing data inside

```python
data_chemin='data/spam.csv'
data=pd.read_csv(data_chemin)
```

```python
data
```

```python
s=data['Category'].value_counts()
print(s)
```

```python
data.shape
```

```python
data['spam'] = data['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
data.shape
```

```python
data.head()
```

```python
data.groupby('Category').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```

```python
data.drop('Category', inplace = True,axis=1)
```

```python
data.tail()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
k_fold = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
X = data['Message']
y = data['spam']
```

```python
for train_index, test_index in k_fold.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
data = CountVectorizer()

X_train_cv = data.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.shape
```

```python
data.get_feature_names_out()[1770]
```

```python
data.vocabulary_
```

```python
model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = data.transform(X_test)
```

```python
y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 7))
plt.hist([y_test, y_pred], label=['Actual', 'Predicted'])
plt.legend(loc='upper right')

plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Actual vs. Predicted Spam Distribution')
plt.show()
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Spam', 'Spam'], yticklabels=['Not Spam', 'Spam'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
#print("Confusion Matrix:")
#print(conf_matrix)
```
