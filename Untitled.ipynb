importing libraries and reading data from the dataset 

```python
import pandas as pd
import numpy as np
```

```python
data2_chemin='spam.csv'
data2=pd.read_csv(data2_chemin)
```

```python
data2
```

```python
s=data2['Category'].value_counts()
print(s)
```

```python
data2.shape
```

```python
data2['spam'] = data2['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
data2.shape
```

```python
data2.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data2.Message, data2.spam, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

data = CountVectorizer()

X_train_cv = data.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.shape
```

```python
data.get_feature_names_out()[1770]
```

```python
data.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = data.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.hist([y_test, y_pred], label=['Actual', 'Predicted'])
plt.legend(loc='upper right')
plt.show()
```
